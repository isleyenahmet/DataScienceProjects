```python
# Veri Seti Hikayesi
```

```python
import seaborn as sns
fmri = sns.load_dataset("fmri")
df = fmri.copy()
df.head()
```

```python
#Beyne bağlanan cihaz aracılığı ile toplanan sinyalleri ifade eden bir veri seti
```

```python
df["timepoint"].describe()
```

```python
df["signal"].describe()
```

```python
df.groupby("timepoint")["signal"].count() #timepointleri gruplandır ve sinyalleri göster
```

```python
df.groupby("signal")["timepoint"].count() #o zaman her bir zaman noktası de eşsizdir
```

```python
df.groupby("timepoint")["signal"].describe() #zaman noktalarına göre sinyallerin degerlerini inceleyelim
```

```python
sns.lineplot(x = "timepoint", y = "signal",hue = "event", data = df);
```

```python
sns.lineplot(x = "timepoint", y = "signal",hue = "event",style = "event" ,data = df);
```

```python
sns.lineplot(x = "timepoint", y = "signal",hue = "event",style = "event" ,markers = True, dashes = False, data = df); #Ortalama Degerleri İşaretledik
```

```python
sns.lineplot(x = "timepoint", y = "signal",hue = "region",style = "event" ,markers = True, dashes = False, data = df); #Ortalama Degerleri İşaretledik
```
