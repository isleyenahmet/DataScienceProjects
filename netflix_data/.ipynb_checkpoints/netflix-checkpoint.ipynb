# Grafiklerle İnceleyelim

```python
#Kullanıcı Yaşlarının Dağılımını İnceleyelim
```

```python
sns.histplot(df["Age"], kde= True);
```

```python
#Ülkeye Göre Kullanıcı Sayisi
```

```python
sns.countplot(x = "Country", data = df);
```

```python
#Abonelik Türüne Göre Dağılım
```

```python
df["Subscription_Type"].value_counts().plot.pie(autopct='%1.1f%%');
```

```python
#Hangi Türler Daha Çok Tercih Ediliyor?
```

```python
sns.countplot(y="Favorite_Genre", data = df, order = df["Favorite_Genre"].value_counts().index);
```

```python
#Abonelik Türüne Göre Ortalama İzlenme
```

```python
sns.boxplot(x = "Subscription_Type", y = "Watch_Time_Hours", data=df);
```

```python
#Hangi Yaş Grubu Daha Fazla İzliyor?
```

```python
sns.scatterplot(x="Age", y="Watch_Time_Hours", hue="Subscription_Type",data=df);
```

```python
#Hangi Abonler Hangi Tür İçeriği Daha Fazla İzliyor?
```

```python
sns.countplot(x = "Favorite_Genre", hue= "Subscription_Type", data = df);
```

```python
#Hangi Ülke Ne Tarz İçerik Tüketiyor?
```

```python
sns.countplot(x = "Favorite_Genre", hue= "Country", data = df);
```
