| Kolon Adı     | Açıklama                                                                                                                      |
| ------------- | ----------------------------------------------------------------------------------------------------------------------------- |
| **age**       | Müşterinin yaşı (sayı değeri).                                                                                                |
| **job**       | Müşterinin mesleği (örneğin: technician, retired, student vb.).                                                               |
| **marital**   | Medeni durumu (single, married, divorced).                                                                                    |
| **education** | Eğitim durumu (primary, secondary, tertiary, unknown).                                                                        |
| **default**   | Kredi temerrüt durumu (önceden kredisi olup geri ödeyememe durumu) — `yes` veya `no`.                                         |
| **balance**   | Banka hesabındaki mevcut bakiye (Euro cinsinden).                                                                             |
| **housing**   | Konut kredisi olup olmadığı — `yes` veya `no`.                                                                                |
| **loan**      | Kişisel kredi borcu olup olmadığı — `yes` veya `no`.                                                                          |
| **contact**   | Kampanya süresince kullanılan iletişim türü (cellular, telephone, unknown).                                                   |
| **day**       | Son iletişimin gerçekleştiği gün (ayın kaçı).                                                                                 |
| **month**     | Son iletişimin gerçekleştiği ay (örneğin: may, nov).                                                                          |
| **duration**  | Son iletişimin süresi (saniye cinsinden). Bu değer, hedef değişkeni etkilediği için modellemede dikkatli kullanılmalıdır.     |
| **campaign**  | Bu müşteriyle yapılan iletişim sayısı (bu kampanyada).                                                                        |
| **pdays**     | Müşteriyle son iletişimden bu yana geçen gün sayısı (önceki kampanyada). `-1` değeri, hiç iletişim kurulmamış anlamına gelir. |
| **previous**  | Önceki kampanyalarda müşteriyle yapılan iletişim sayısı.                                                                      |
| **poutcome**  | Önceki pazarlama kampanyasının sonucu (success, failure, unknown, other).                                                     |
| **deposit**   | Hedef değişken — Müşteri vadeli mevduat (term deposit) açtı mı? (`yes` veya `no`)                                             |
