# 4-Kategorik Verileri Sayılara Dönüştür

```python
df["Sex"] = df["Sex"].map({"male" : 0 , "female": 1})
df["Embarked"] = df["Embarked"].map({"S": 0, "C": 1, "Q": 2}) #Burada verileri makinenin anlayabileceği degerlere donusturduk
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
#Özellikleri belirleyelim:
```

```python
x = df[["Pclass", "Sex","Age","SibSp","Parch","Fare","Embarked",]]
y = df["Survived"]
```

```python
# Veriyi Eğitim ve Test olarak ayıralım
```

```python
x_train , x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
```

```python
# Modeli tanımlama ve eğitme kısmı
```

```python
model = RandomForestClassifier() # Birden fazla karar ağacı oluşturup çoğunluk kararına göre sınfılandırma yapar
model.fit(x_train, y_train) #Öretir yani eğitim yapar
```

```python
#Tahmin Yaptıralım
```

```python
predictions  = model.predict(x_test)
```

```python
predictions
```

```python
#Model yukarıda tahminlerini yaptı ve y ile de şimdi doğruluk oranını kiyaslayacağız
```

```python
print("Dogruluk: ", accuracy_score(y_test, predictions))
```
