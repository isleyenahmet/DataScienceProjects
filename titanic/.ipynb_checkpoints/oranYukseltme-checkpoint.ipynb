```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users/isley/Documents/GitHub/DataScienceProjects/titanic/train.csv")
```

```python
df.isnull().sum()
```

```python
df["Age"].fillna(df["Age"].mean(), inplace=True) #Ortalama değer ile doldurdum
```

```python
df.isnull().sum() # tekrardan kontrol ettim
```

```python
df.drop(columns=["Cabin"], inplace = True)
```

```python
df.isnull().sum()
```

```python
df["Embarked"].fillna(df["Embarked"].mode()[0], inplace = True) #burada da nereden bindikleri bilgisi 2 kişide eksikti ve bunları da en fazla binilen yerden telafi ettik
```

```python
df.isnull().sum()
```

```python
df["Sex"] = df["Sex"].map({"male" : 0 , "female": 1})
df["Embarked"] = df["Embarked"].map({"S": 0, "C": 1, "Q": 2}) #Burada verileri makinenin anlayabileceği degerlere donusturduk
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
x = df[["Pclass", "Sex","Age","SibSp","Parch","Fare","Embarked",]]
y = df["Survived"]
```

```python
x_train , x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
```

```python
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
```

```python
xgb_model.fit(x_train, y_train)
```

```python
xgb_preds = xgb_model.predict(x_test)
```

```python
print("XGBoost Doğruluk:", accuracy_score(y_test, xgb_preds))
```
